```python
import numpy as np
import matplotlib.pyplot as plt
import math

mb=1
D=10**-3
pi=3.14
t=100
z=np.arange(-3,3,0.01)
y=[]
for i in z:
    val=mb*math.exp(-i**2/(4*D*t))/math.sqrt(4*pi*D*t)
    y.append(val)
    if val>=0.45 and val <= 0.55:
      print(i,val)
plt.plot(z,y)
plt.show()
```

```python
Cs=10
Co=5
D=10**-3
t=100
x=np.arange(0,2,0.01)
y=[]
for i in x:
    val=mb*math.erf(i/math.sqrt(4*D*t))*(Co-Cs)+Cs
    y.append(val)
    if val>=5.0 and val <= 5.02:
      print(i,val)
plt.plot(x,y)
plt.show()
```

```python
y=[np.concatenate((np.ones(len(np.arange(0,2,0.1))),np.zeros(len(np.arange(2,4,0.1))))),]
dx=0.1
dt=0.1
D=0.02
s=D*dx/(dt**2)
s=0.2
Cm=np.concatenate((np.ones(len(np.arange(0,2,0.1))),np.zeros(len(np.arange(2,4,0.1)))))
for i in np.arange(0.1,2.6,0.1): # 0ÏùÄ initial condition
  tmp=np.zeros(len(np.arange(0,4,0.1)))
  tmp[0]=1
  idx=1
  for j in np.arange(0.1,3.9,0.1):
    tmp[idx]=(1-2*s)*Cm[idx]+s*(Cm[idx+1]+Cm[idx-1])
    idx+=1
  y.append(tmp)
  Cm=tmp
plt.plot(np.arange(0,4,0.1),y[15])
plt.show()
```

```python
pD=5*10**-6
pT=500
dD=2*10**-5
dT=200
s=10**19
#step x -> 5*10**-4
x=0
C=s*math.erfc(x/(2*math.sqrt(pD*pT)))
while C >= 10**14:
  x+=5*10**-4
  C=s*math.erfc(x/(2*math.sqrt(pD*pT)))
  print(C,x)
```

```python
pD=5*10**-6
pT=500
dD=2*10**-5
dT=200
s=10**19
#step x -> 5*10**-4
x=0
C=(2*s/pi)*((pD*pT)/(dD*dT))*math.exp(-x**2/(4*dD*dT))
while C >= 10**14:
  x+=5*10**-4
  C=(2*s/pi)*((pD*pT)/(dD*dT))*math.exp(-x**2/(4*dD*dT))
  print(C,x)
```

```python
Q=10**13
Rp=0.1*10**-4
dRp=0.04*10**-4
D=6.5*10**-13
t=60
C=np.zeros(len(np.arange(0,0.00011,0.00000001)))
N=np.zeros(len(np.arange(0,0.00011,0.00000001)))
idx=0
for i in np.arange(0,0.00011,0.00000001):
  C[idx]=Q*math.exp(-(i-Rp)**2/(2*dRp**2))/(math.sqrt(2*pi)*dRp)
  N[idx]=Q*math.exp(-(i-Rp)**2/((dRp)**2+2*D*t))/(math.sqrt(2*pi)*math.sqrt((dRp)**2+2*D*t))
  idx+=1
plt.plot(np.arange(0,0.00011,0.00000001),C[0:11000],label='implantation')
plt.plot(np.arange(0,0.00011,0.00000001),N[0:11000],label='annealing')
plt.legend()
plt.show()
```
